```python
# Import required modules
import dns_property as dns
import math
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import requests
import seaborn as sns
import warnings

from numpy.random import seed
from lightgbm import LGBMRegressor
from os import listdir
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics import r2_score, mean_squared_error, mean_absolute_error
from sklearn.model_selection import RepeatedKFold, train_test_split
from sklearn.neighbors import KNeighborsRegressor
from sklearn.preprocessing import MinMaxScaler, LabelEncoder

# Settings
warnings.filterwarnings('ignore')
```

## Import Data

```python
# Read data
df = pd.read_csv('../Data/resale_final.csv')

# Preview
df.head()
```

## Feature Engineering

```python
# Compute floor percentage
df['floor_pct'] = df.floor / df.max_floor_lvl

# Create month
df['month'] = pd.to_datetime(df['month'])
df = df.rename(columns={'month': 'mth'})
```

## Conventional CV

```python
df.multistorey_carpark.value_counts()
```

```python
for i in df.columns:
    if df[i].nunique() < 5:
        print(df[i].value_counts())
        print()
```

```python
# HDB Vars
vars_hdb_drop = ['target', 'mth', 'block_letter']
vars_hdb_bin = ['town', 'flat_type', 'flat_model', 'block_num']

# Configure data
X_data = pd.get_dummies(df.drop(vars_hdb_drop, axis=1), columns=vars_hdb_bin)
y_data = df['target']

# Configure CV
cv = RepeatedKFold(n_splits=5, n_repeats=1, random_state=100)

# Configure stop words
stop_words = [
    'road', 'avenue', 'street', 'drive', 'lorong', 'crescent', 'view', 'rise'
    'jalan', 'lane', 'walk', 'link', 'bukit',
    'rd', 'ave', 'st', 'dr', 'lor', 'cres', 'jln', 'pl'
]
```

### K-NN
* With `block_letter`: 4.00% PPD
* Without `block_letter`: 3.94% PPD

```python
kn = KNeighborsRegressor(n_neighbors=10, weights='distance', n_jobs=4)
kn_cv = dns.custom_cv(kn, X_data, y_data, cv, ['street_name'], stop_words, norm=True, ppd=True)
```

### LightGBM
Feature selection. If PPD drops, backtrack to previous best and continue:
* With `block_letter`: 3.18% PPD
* Without `block_letter`: 3.16%
* Without `flat_model`: 3.18%
* Removed low variation features: 3.17%

```python
lg = LGBMRegressor(n_estimators=5000, max_depth=25, random_state=123, n_jobs=4, reg_alpha=0.1, reg_lambda=0.9)
lg_cv = dns.custom_cv(lg, X_data, y_data, cv, ['street_name'], stop_words, ppd=True)
```

## Step-Forward Testing

```python
# Private Vars
vars_hdb_drop = ['target']
vars_hdb_bin = ['town', 'flat_type', 'flat_model', 'block_num', 'block_letter']

# Configure data
X_data = pd.get_dummies(df.drop(vars_hdb_drop, axis=1),
                        columns=vars_hdb_bin)
y_data = df['target']

# Configure stop words
stop_words = [
    'road', 'avenue', 'street', 'drive', 'lorong', 'crescent', 'view', 'rise'
    'jalan', 'lane', 'walk', 'link', 'bukit',
    'rd', 'ave', 'st', 'dr', 'lor', 'cres', 'jln', 'pl'
]
```

### K-NN

```python
kn = KNeighborsRegressor(n_neighbors=10, weights='distance', n_jobs=4)
kn_cv = dns.custom_ts(rgr=kn, X=X_data, y=y_data, timevar='mth', start='2018-09-01', tfidf_cols=['street_name'], stop_words=stop_words, dropout=0.1, ppd=True, random_state=100)
```

### LightGBM

```python
lg = LGBMRegressor(n_estimators=5000, max_depth=25, random_state=123, n_jobs=4, reg_alpha=0.1, reg_lambda=0.9)
lg_cv = dns.custom_ts(rgr=lg, X=X_data, y=y_data, timevar='mth', start='2018-09-01', tfidf_cols=['street_name'], stop_words=stop_words, dropout=0.1, ppd=True, random_state=100)
```

## Step-Forward Testing with FE

```python
# Private Vars
vars_hdb_drop = ['target', 'block_letter', 'mth_l', 'qtr', 'qtr_l']
vars_hdb_bin = ['town', 'flat_type', 'flat_model', 'block_num']

# Copy data
df2 = df.copy()

# Generate month lags
for var in ['town', 'flat_type', 'flat_model', 'block_num', 'street_name']:

    df2 = dns.fe_mth(df2, var, 1)
    df2 = dns.fe_qtr(df2, var, 1)

# Configure data
X_data = pd.get_dummies(df2.drop(vars_hdb_drop, axis=1),
                        columns=vars_hdb_bin)
y_data = df2['target']

# Configure stop words
stop_words = [
    'road', 'avenue', 'street', 'drive', 'lorong', 'crescent', 'view', 'rise'
    'jalan', 'lane', 'walk', 'link', 'bukit',
    'rd', 'ave', 'st', 'dr', 'lor', 'cres', 'jln', 'pl'
]
```

### K-NN

```python
kn = KNeighborsRegressor(n_neighbors=10, weights='distance', n_jobs=4)
kn_cv = dns.custom_ts(rgr=kn, X=X_data, y=y_data, timevar='mth', start='2018-09-01', tfidf_cols=['street_name'], stop_words=stop_words, dropout=0.1, ppd=True, random_state=100)
```

### LightGBM

```python
lg = LGBMRegressor(n_estimators=5000, max_depth=25, random_state=123, n_jobs=4, reg_alpha=0.1, reg_lambda=0.9)
lg_cv = dns.custom_ts(rgr=lg, X=X_data, y=y_data, timevar='mth', start='2018-09-01', tfidf_cols=['street_name'], stop_words=stop_words, dropout=0.1, ppd=True, random_state=100)
```

## Step-Forward Testing with FE+

```python
# Private Vars
vars_hdb_drop = ['target', 'block_letter', 'mth_l', 'qtr', 'qtr_l']
vars_hdb_bin = ['town', 'flat_type', 'flat_model', 'block_num']

# Copy data
df2 = df.copy()

# Generate month lags
for var in ['town', 'flat_type', 'flat_model', 'block_num', 'street_name']:

    df2 = dns.fe_mth(df2, var, 1)
    df2 = dns.fe_qtr(df2, var, 1)
    df2 = dns.fe_qtr(df2, var, 2)

# Configure data
X_data = pd.get_dummies(df2.drop(vars_hdb_drop, axis=1),
                        columns=vars_hdb_bin)
y_data = df2['target']

# Configure stop words
stop_words = [
    'road', 'avenue', 'street', 'drive', 'lorong', 'crescent', 'view', 'rise'
    'jalan', 'lane', 'walk', 'link', 'bukit',
    'rd', 'ave', 'st', 'dr', 'lor', 'cres', 'jln', 'pl'
]
```

### K-NN

```python
kn = KNeighborsRegressor(n_neighbors=10, weights='distance', n_jobs=4)
kn_cv = dns.custom_ts(rgr=kn, X=X_data, y=y_data, timevar='mth', start='2018-09-01', tfidf_cols=['street_name'], stop_words=stop_words, dropout=0.1, ppd=True, random_state=100)
```

### LightGBM
* With 2 quarter lag, no difference, 3.99%

```python
lg = LGBMRegressor(n_estimators=5000, max_depth=25, random_state=123, n_jobs=4, reg_alpha=0.1, reg_lambda=0.9)
lg_cv = dns.custom_ts(rgr=lg, X=X_data, y=y_data, timevar='mth', start='2018-09-01', tfidf_cols=['street_name'], stop_words=stop_words, dropout=0.1, ppd=True, random_state=100)
```
